```python
import re

def is_valid_date(date_string):
    # Define a regular expression pattern for the "yyyy-mm-dd" format
    date_pattern = r'^\d{4}-(0[1-9]|1[0-2])-(0[1-9]|[12][0-9]|3[01])$'

    # Use the re.match function to check if the input matches the pattern
    if re.match(date_pattern, date_string):
        # Split the date into year, month, and day components
        year, month, day = map(int, date_string.split('-'))

        # Check if the year is within a reasonable range (e.g., 1900-2099)
        if 1900 <= year <= 2023:
            # Check if the month is valid (1-12)
            if 1 <= month <= 12:
                # Check if the day is valid for the given month
                if 1 <= day <= 31:
                    # Additional checks can be added, e.g., February and leap years
                    return True
    return False

# Test cases
print(is_valid_date("2023-10-19"))  # Valid date format and valid values
print(is_valid_date("2023-13-19"))  # Invalid month (13 is not a valid month)
print(is_valid_date("2023-02-30"))  # Invalid day for February (30 is not valid)
print(is_valid_date("1899-12-31"))  # Invalid year (below the reasonable range)
```

```python
df['date'] = df['date'].apply(lambda x: x if is_valid_date(x) else np.nan)
```
